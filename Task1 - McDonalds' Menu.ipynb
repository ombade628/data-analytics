### <b> Visualize distribution of 'Calories'</b>
- to identify any skewed data or outliers

```python
plt.figure(figsize=(8, 6))
sns.histplot(df['Calories'], bins=20, kde=True)
plt.title('Calories Distribution')
plt.xlabel('Calories')
plt.ylabel('Frequency')
plt.show()
```

### <b>  Box plot for Total Fat across different categories</b>

```python

plt.figure(figsize=(12, 6))
sns.boxplot(data=df, x='Category', y='Total Fat', palette='Set3')
plt.title('Box Plot of Total Fat by Category')
plt.xlabel('Category')
plt.ylabel('Total Fat (g)')
plt.xticks(rotation=45)
plt.show()
```

### <b> Box plot for Sodium across different categories </b>

```python
plt.figure(figsize=(12, 6))
sns.boxplot(data=df, x='Category', y='Sodium', palette='Set2')
plt.title('Box Plot of Sodium by Category')
plt.xlabel('Category')
plt.ylabel('Sodium (mg)')
plt.xticks(rotation=45)
plt.show()
```

## 4. EDA

### <b> Category Analysis: Average Calories per Category</b>

```python
plt.figure(figsize=(10, 6))
category_avg_calories = df.groupby('Category')['Calories'].mean().sort_values()
category_avg_calories.plot(kind='barh', color='skyblue')
plt.title('Average Calories per Category')
plt.xlabel('Calories')
plt.ylabel('Category')
plt.show()
```

> Insights - 
1. `Non-veg` has <b>High Calories</b> followed by `Smoothies & Shakes` and `Breakfast`.
2. `Beverages` has <b>Low Claories</b>.


### <b> Nutritional Analysis - Fat and Carbohydrates

```python
plt.figure(figsize=(12, 8))
sns.scatterplot(data=df, x='Total Fat', y='Carbohydrates', hue='Category')
plt.title('Total Fat vs Carbohydrates by Category')
plt.xlabel('Total Fat (g)')
plt.ylabel('Carbohydrates (g)')
plt.legend(loc='upper right')
plt.show()
```

> Insights - 
1. `Smoothies & Shakes` have <b>High Carbohydrates</b>.
2. `Breakfast` includes <b>High Fat and Carbohydrates</b>.
3. `Salads` have <b>Low Fat and Carbohydrates</b>.

### <b> Correlation Heatmap

```python
# Select only the numeric columns from the DataFrame
numeric_df = df.select_dtypes(include=['float64', 'int64'])

# Calculate the correlation matrix on the numeric data
correlation_matrix = numeric_df.corr()

# Plot the heatmap
plt.figure(figsize=(14, 8))  # Set the figure size
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt='.2f', linewidths=0.5)
plt.title('Correlation Heatmap Matrix')
plt.show()
```
